```python
def generate(filename, imagename):
  #import necessary packages
  import numpy as np
  import pandas as pd
  from wordcloud import WordCloud,STOPWORDS,ImageColorGenerator
  import matplotlib.pyplot as plt
  from PIL import Image

  #read the file
  text = open(filename).read()
  #generate the WordCloud
  wordcloud = WordCloud().generate(text)
  #display the WordCloud
  plt.imshow(wordcloud)
  plt.axis("off")
  plt.show()
  #open the user given image
  hatmask = np.array(Image.open(imagename))
  #display the image
  plt.imshow(hatmask)
  plt.axis("off")
  plt.show()
  #create the list of stop words
  stopwords = set(STOPWORDS)
  wordcloud = WordCloud(background_color="white",
  mask=hatmask,contour_width=3,
  stopwords = STOPWORDS,
  repeat=True,
  min_font_size=3,contour_color="darkgreen")

  #generate the final wordcloud
  wordcloud.generate(text)
  #show it to the user
  plt.imshow(wordcloud)
  plt.axis("off")
  plt.show()
```

```python
#showing the user all of the given WordClouds
generate("Hackathon.txt","hatbruvv.png")
generate("winstonchurchillspeech.txt","winstonchurchill.jpg")
generate("lebron.txt","lebronpic.png")
```
