```python
#IMPORT PACKAGES
import numpy as np
import plotly.graph_objects as go
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import seaborn as sns

def wc(file):
  #OPEN TEXT FILE
  text = open(file).read()
  #SPLIT INTO A LIST BY SPACES
  tokens_without_sw = text.split()
  print(tokens_without_sw)


  def common(List):
  #RETURN THE MOST COMMON WORD IN THE GENERATED LIST
    count = 0
    number = List[0]

    for i in List:
        count_frequency = List.count(i)
        if(count_frequency> count):
            count = count_frequency
            number = i

    return number


  print(common(tokens_without_sw))


  stopwords = [',', '.', 'I', '?', ';', '-','the','you',"I'm",'your',"don't", 'He','he','like','was','would','an','into','very','and', 'for', 'of', 'to','be','in','a', 'is','had','by','not','it','or','as','all','no','any','can','will','with','are','have','this','which','so','you','that','from','upon','do','there','their','such','but','shall','than','they','on','may','its','if','while','who','It','If','has','these','In','at','The','You',"I'm",'your','we',"don't","It's","What",'what','just','my']
  # ^ WORDS WHCIH DO NOT ADD VALUE TO THE TEXT, NEED TO BE REMOVED
  for x in stopwords:
    count4 = tokens_without_sw.count(x)
    for y in range(count4):
      #REMOVE FROM THE LIST
      tokens_without_sw.remove(x)

  print(tokens_without_sw)
  print(common(tokens_without_sw))

  #CREATE A DICTIONARY SO THAT WORD:COUNT PAIRS CAN BE STORED
  dictword = {}
  #CREATE A COPY LIST, CANNOT EDIT MASTER LIST
  new_list = tokens_without_sw.copy()

  #LIMIT LIST TO 30 ELEMENTS (30 MOST COMMON WORDS)
  for x in range(31):
    most_common = common(new_list)
    most_count = new_list.count(most_common)
    dictword[most_common] = most_count
    for y in range(most_count):
      while most_common in new_list: new_list.remove(most_common)


  print(dictword)

  #VISUAL SCATTER PLOT
  from pandas import DataFrame
  d = {'Word':dictword.keys(), 'Count': dictword.values()}
  df = DataFrame(data=d)
  df.head()

  size = [90, 75]
  fig = go.Figure(data=[go.Scatter(
    x=df['Word'],
    y=df['Count'],
    text=df['Word'],
    mode='markers+text',
    name='Markers and Text',
    textposition='top center',
    marker=dict(
        size=df['Count']*10,
        sizemode='area',
        sizeref=2.*max(size)/(40.**2),
        sizemin=4,
    )
  )])
  fig.update_xaxes(visible=False, showticklabels=False)
  fig.update_layout(width=2000)
  fig.show()

```

```python
wc("Hackathon.txt")
wc("winstonchurchillspeech.txt")
wc("lebron.txt")
```
